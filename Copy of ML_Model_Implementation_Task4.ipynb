```python
import pandas as pd

#  Sample data similar to the SMS Spam Collection dataset
data = {
    'v1': ['ham', 'spam', 'ham', 'spam', 'ham'],
    'v2': [
        'Hey, are we still meeting for lunch today?',
        'WINNER!! You’ve won a $1000 Walmart gift card. Click to claim.',
        'Can you send me the notes from class?',
        'Congratulations! You’ve been selected for a FREE vacation.',
        'I’ll call you back in 5 minutes.'
    ]
}

#  Create DataFrame
df = pd.DataFrame(data)

#  Save as spam.csv
df.to_csv('spam.csv', index=False, encoding='utf-8')

print("spam.csv created successfully!")
```

```python
import pandas as pd

# Load the spam.csv you just created
df = pd.read_csv("spam.csv")

# Rename columns for clarity
df.columns = ['label', 'message']

# Show the first 5 messages
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# Convert labels ('ham'/'spam') to numbers (0/1)
le = LabelEncoder()
df['label_num'] = le.fit_transform(df['label'])

# Split dataset
X = df['message']
y = df['label_num']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Convert text into TF-IDF vectors
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Train Naive Bayes model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Predict on test set
y_pred = model.predict(X_test_vec)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Full classification report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, target_names=['Ham', 'Spam']))
```

```python
# Try your own message
your_msg = ["You have WON a free laptop! Click here to claim"]
your_msg_vec = vectorizer.transform(your_msg)
predicted = model.predict(your_msg_vec)

print(" Prediction:", "SPAM" if predicted[0] == 1 else "HAM")
```
